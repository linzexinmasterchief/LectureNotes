# 14 Partial Derivatives

## <span style="color:#005577">14.1 Functions of Several Variables</span>

$$w=f\left(x_{1}, x_{2}, \ldots, x_{n}\right)$$

> **Domain** $\it{D}:=\{x_1,x_2,...,x_n\}$  
**real-valued function** $\it{f}$  
**Range** $\it{R}:=\{w_1,w_2,...,w_n\}$

*w*: **dependent** variable of *f*  
*f*: function of the n **independent** variables $x_1$ to $x_n$.  
*x_j*: **input** variables of *f*  
*w*: **output** variables of *f*

### <span style="color:#00aaee">14.1.1 Domains and ranges</span>

Exlude inputs that lead to complex numbers / division by zero  

**Domain**: largest set generates **real** numbers  
**Range**: set of **output** for the dependent variable

<span style="color:red">**EXAMPLE**</span>

**(a)** functions of two variables

$
\begin{array}{lll}
\hline \text { Function } & \text { Domain } & \text { Range } \\
\hline z=\sqrt{y-x^{2}} & y \geq x^{2} & {[0, \infty)} \\
z=\frac{1}{x y} & x y \neq 0 & (-\infty, 0) \cup(0, \infty) \\
z=\sin x y & \text { Entire plane } & {[-1,1]} \\
\hline
\end{array}
$

**(b)** functions of three variables

$
\begin{array}{lll}
\hline \text { Function } & \text { Domain } & \text { Range } \\
\hline w=\sqrt{x^{2}+y^{2}+z^{2}} & \text { Entire space } & {[0, \infty)} \\
w=\frac{1}{x^{2}+y^{2}+z^{2}} & (x, y, z) \neq(0,0,0) & (0, \infty) \\
w=x y \ln z & \text { Half-space } z>0 & (-\infty, \infty) \\
\hline
\end{array}
$

Closed intervals [ *a*, *b* ] **include** boundary points  
Opend intervals ( *a*, *b* )  
Neither [ *a*, *b* )

**Interior point**:  
$p(x_0,y_0)$ in a region(set) $R$ in the *xy*-plane **if** it is a center of a **disk** of **positive radius** that lies entirely in $R$  
  
**Boundary point**:  
$p(x_0,y_0)$ ... **if** *every* disk centered at $(x_0,y_0)$ contains points that lie outside of $R$ **as well as** points that lie in $R$  
  
Set of  
**interior** points make up the **interior** of the region  
... **boundary** ... **boundary** ...  
  
A set is  
**Open** if consists entirely of interior points  
**Close** ... boundary points

Region is  
**bounded**:  
lies inside a disk of finite radius.  
*triangles, rectangles, circles...* 

**unbounded**:  
NOT bounded
$\overline{\text {Bounded}}$  
*lines, coordinate axes...*

<span style="color:red">**EXAMPLE**</span>  
Domain of $f(x, y)=\sqrt{y-x^2}$.  
$y-x^2\ge0$, closed, unbounded domain

### <span style="color:#00aaee">14.1.2 Functions of Two Variables</span>

**level curve**: Contour plot  
**graph** = **surface $z=f(x,y)$**

```python
# example 2 variable function
fig = plt.figure()
ax = fig.gca(projection='3d')
 
x = np.arange(-10,10,0.1)
y = np.arange(-10,10,0.1)
X,Y = np.meshgrid(x, y)
Z = 100-X**2-Y**2

ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)

ax.set_zlim(-100, 100)
plt.show()
```

$f(x,y)=100-x^2-y^2   \uparrow$  
Domain: entire $xy$-plane  
Range: real number set $R\le100$  
Level surface: $f(x,y)=a$, $a$ is a constant  
$f(x,y)=100$ consists the origin alone (still a level curve)

## <span style="color:#005577">14.2 Limits and Continuity in Higher Dimensions</span>

### <span style="color:#00aaee">14.2.1 Limits for Functions of Two Variables</span>

$$\lim _{(x,y)\rightarrow(x_0,y_0)} f(x,y)=L$$  
$|f(x,y)-L|\le \epsilon$ whenever $0\lt \sqrt{(x-x_0)^2+(y-y_0)^2} \lt \epsilon$

$$\lim _{(x,y)\rightarrow(x_0,y_0)} x=x_0$$
$$\lim _{(x,y)\rightarrow(x_0,y_0)} y=y_0$$
$$\lim _{(x,y)\rightarrow(x_0,y_0)} k=k$$

$0\lt \sqrt{(x-x_0)^2+(y-y_0)^2} \lt \epsilon$  
implies  
$\sqrt{(x-x_0)^2} \lt \epsilon \rightarrow |x-x0| \lt \epsilon \rightarrow |f(x,y)-x_0| \lt \epsilon$
